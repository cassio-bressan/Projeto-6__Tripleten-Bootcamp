```python
# SUBSTITUINNDO OS NOMES DAS COLUNAS POR M√çNUSCULOS.
def lowercase_columns(df):
    new_columns = [col.lower() for col in df.columns]
    df.columns = new_columns
    return df


df_games = lowercase_columns(df_games)
print(df_games.columns)
```

```python
# SUBSTITUINDO OS DADOS PARA MIN√öSCULO
columns_lower = ['name', 'platform', 'genre']
for data in columns_lower:
    df_games[data] = df_games[data].str.lower()
```

### TESTANDO HIP√ìTESES

```python
xone_score = df_games[df_games['platform'] == 'xone']['user_score'].dropna()
pc_score = df_games[df_games['platform'] == 'pc']['user_score'].dropna()

xone_score.var(), pc_score.var()
```

```python
# Teste as hip√≥teses
# Hipotese nula - as classifica√ß√µes m√©dias dos usuarios das plataformas xbox one e pc s√£o as mesmas.
# Hipotese alternativa - as classifica√ß√µes m√©dias dos usuarios das plataformas xbox one e pc s√£o diferentes
alpha = 0.05

results = st.ttest_ind(xone_score, pc_score, equal_var=False)

print('valor-p:', results.pvalue)

if results.pvalue < alpha:
    print("Rejeitamos a hip√≥tese nula. As m√©dias dos usu√°rios das duas plataformas s√£o significativamente diferentes.")
else:
    print("N√£o rejeitamos a hip√≥tese nula. N√£o h√° diferen√ßa significativa entre as m√©dias dos usu√°rios das duas plataformas.")
```

``Nesse teste de hip√≥tese, podemos verificar que a classifica√ß√£o m√©dia dos usu√°rios para duas diferentes plataformas s√£o diferentes. Isso significa que n√£o podemos afirmar nada para x plataforma tomando como base os dados de outra plataforma completamente diferente. Provavelmente a imers√£o, gr√°ficos e desempenho dos jogos √© diferente nas duas plataformas, e por isso a classifica√ß√£o dos usu√°rios tamb√©m √© diferente.``

```python
action_score = df_games[df_games['genre'] == 'action']['user_score'].dropna()
sports_score = df_games[df_games['genre'] == 'sports']['user_score'].dropna()

action_score.var(), sports_score.var()
```

```python
# Teste as hip√≥teses
# Hipotese nula - as classifica√ß√µes m√©dias dos usuarios dos g√™neros action e sports s√£o as mesmas.
# Hipotese alternativa - as classifica√ß√µes m√©dias dos usuarios dos g√™neros action e sports s√£o diferentes
alpha = 0.05

results = st.ttest_ind(action_score, sports_score, equal_var=False)

print('valor-p:', results.pvalue)

if results.pvalue < alpha:
    print("Rejeitamos a hip√≥tese nula. As m√©dias dos usu√°rios para os dois g√™neros s√£o significativamente diferentes.")
else:
    print("N√£o rejeitamos a hip√≥tese nula. N√£o h√° diferen√ßa significativa entre as m√©dias dos dois g√™neros.")
```

``Nesse teste de hip√≥tese, n√£o podemos rejeitar a hip√≥tese nula. Podemos verificar que as classifica√ß√µes m√©dias dos usu√°rios para os dois g√™neros n√£o possuem uma diferen√ßa significativa, e com isso podemos levantar a hip√≥tese de que os dois g√™neros possuem tantos jogadores fi√©is √† sua jogabilidade individual, que a m√©dia de seus usu√°rios se torna ligeiramente similar.``

`` - Para formular as hip√≥teses nulas e alternativas, tomo como base que a nula √© a hip√≥tese que eu quero testar, enquanto a alternativa √© a contra resposta caso a nula seja rejeitada.``
